## Data Cleaning

First, in pricing column some values are entered as POA (which stands for Price on Application). That means the seller of the property will only discuss the price of the property to the customer instead of making the property price publicly available [link](https://www.property24.com/articles/poa-what-does-it-mean-and-why-do-sellers-use-it/29977).

```python
 # Number of customers who tagged the price as POA.
len(df[df["price"]=="POA"])
```

```python
# We will remove properties with POA on price column from here
df = df[df["price"]!="POA"]
```

```python
# Other properties also does not have any values. We have already seen that there are 532 records. Let's remove this.
df = df[df["price"].notnull()]
```

```python
def split_price_currency(price_str):
    # Currency looks like this: 'R 6 600 000', then we can process as follows:
    # Split the string by the first space, keep the first part (currency), and remove the spaces from the second part (price).
    # Then convert the price to a float and return both the currency and the price.
    parts = price_str.split(' ', 1)
    currency = parts[0]
    price = parts[1].replace(' ', '')
    return currency, price

# Apply the function to the DataFrame and convert it to a pandas Series.
split_cols = df['price'].apply(split_price_currency).apply(pd.Series)
# Separate currency and the numeric part of the price
df.loc[:, 'currency'] = split_cols[0]
df.loc[:, 'price2'] = pd.to_numeric(split_cols[1])
df = df.copy()
```

```python
# check unique currencies.
print(df.currency.unique())
# There is 1 currency: South African Rand, therefore can can just remove the currency column. 
# And the original price column as well.
df.drop(['currency', 'price'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Next, let's clean the Floor Size column. It is measured in mÂ² and precented like '828 mÂ²'
# And there are nan values as well - 7529 entries
print(df["Floor Size"].isnull().sum())
def clean_floor_size(size_str):
    # Remove the last character (mÂ²) and convert it to a float.
    if isinstance(size_str, str):
        return float(size_str.replace(" mÂ²", ""))
    else:
        return np.nan
df["Floor Size"] =df["Floor Size"].apply(clean_floor_size)
```

```python
# There are so many columns with so many values that we may not need at all.
# Let's identify the columns with 24000 missing values out of 20000 expected values and removed them.
columns_with_many_missing = df.columns[df.isnull().sum() > 20000]
print(columns_with_many_missing)
df.drop(columns_with_many_missing, axis=1, inplace=True)
```

```python
# Let's process name column here. It has no missing values.
def extract_address(name_str):
    # name column is structured like this: '4 Bedroom House for Sale in Wilderness Central'
    # We can process as follows:
    # Split the string by the first space, keep the second part (address), and remove the spaces from the rest.
    # Then return the address.
    parts = name_str.split(" in ")
    name, address = parts
    return name, address

# Apply the function to the 'name' column and convert the result to a DataFrame
split_cols_name = df['name'].apply(extract_address).apply(pd.Series)

# Rename the columns appropriately
split_cols_name.columns = ['name', 'address']

# Assign the new columns back to the original DataFrame
df['name'] = split_cols_name['name']
df['address'] = split_cols_name['address']
```

```python
# Next, we convert Listing Date into valid pandas datetime and 
# get the number of days form listing date to the time of data extraction.
df["Listing Date"] = pd.to_datetime(df["Listing Date"], format="%d %B %Y")
df['days_since_listing'] = (datetime(2024,8,7) - df['Listing Date']).dt.days
df.head()
```

```python
# Number of bedrooms and bathrooms have been recorded twice. Let's merge them.

def merge_columns(row, col1, col2):
    # If both columns are NaN, return NaN
    if pd.isna(row[col1]) and pd.isna(row[col2]):
        return np.nan
    # If they are equal, we return one of them.
    if row[col1] == row[col2]:
        return row[col1]
    # if col1 is NaN, return col2 and vice versa
    if pd.isna(row[col1]):
        return row[col2]
    if pd.isna(row[col2]):
        return row[col1]
    return np.nan

df['Bedrooms2'] = df.apply(merge_columns, axis=1, col1='Bedrooms', col2='Bedrooms:')
df['Bathrooms2'] = df.apply(merge_columns, axis=1, col1='Bathrooms', col2='Bathrooms:')
```
