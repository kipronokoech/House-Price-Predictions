```python
df[df["Erf Size"]=="521 acres"]["price"].str.replace(r"[\r\n\s]", "", regex=True).str.strip()
```

```python
df["price"] = df["price"].str.replace(r"[\r\n\s]", "", regex=True).str.strip()
```

```python
df
```

```python
df["currency"] = df["price"].str.extract(r"([^\d]+)")
df["price"] = df["price"].str.extract(r"(\d+)")
df["price2"] = pd.to_numeric(df["price"], errors='coerce')
```

```python
df[["price", "price2", "currency"]]
```

```python
# Convert 'price2' to numeric, setting errors='coerce' to NaN for invalid entries
df["price2_numeric"] = pd.to_numeric(df["price2"], errors='coerce')

# Identify rows where 'price2_numeric' is NaN (indicating invalid entries)
invalid_entries = df[df["price2_numeric"].isna()].price.unique()

print("Rows with invalid 'price2' entries:")
invalid_entries
```
